# Outlier Detection

Outliers are abnormal values.

### Example:

10,12,11,13,15,300

300 = Outlier

## IQR Method
IQR = Q3 - Q1

### Lower Bound:

Q1 - 1.5*IQR

### Upper Bound:

Q3 + 1.5*IQR

## Z-Score  
Z = (X - Mean)/Std

If:

|Z| > 3

Then Outlier
