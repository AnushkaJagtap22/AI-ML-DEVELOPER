```python
from sklearn.cluster import KMeans
import numpy as np

# Sample data: fruit weights and sweetness
X = np.array([
    [150, 7],  # Apple
    [170, 8],  # Apple
    [140, 4],  # Orange
    [160, 5],  # Orange
    [200, 9],  # Mango
    [210, 10]  # Mango
])

# Create KMeans model with k=3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Cluster assignments
print("Cluster labels:", kmeans.labels_)

# Centroids
print("Cluster centers:\n", kmeans.cluster_centers_)

# Predict new fruit
new_fruit = np.array([[155, 6]])  # weight=155, sweetness=6
print("Predicted cluster:", kmeans.predict(new_fruit))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Data (fixed column names)
data = {
    'Customer': ['Anushka', 'Aarush', 'Riya', 'Priya', 'Ram'],
    'Age': [20, 40, 24, 35, 23],
    'Spending': [100, 200, 300, 110, 290]
}

df = pd.DataFrame(data)

# Features
X = df[['Age', 'Spending']]

# Apply KMeans
model = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Group'] = model.fit_predict(X)

# Plot
plt.figure(figsize=(6,5))

for group in df['Group'].unique():
    group_data = df[df['Group'] == group]
    plt.scatter(group_data['Age'], group_data['Spending'], label=f'Group {group}')

plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer Segmentation')
plt.legend()
plt.grid(True)
plt.show()
```
