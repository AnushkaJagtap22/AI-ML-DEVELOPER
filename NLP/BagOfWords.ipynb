```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = [
    'I love pizza',
    'Pizza is the best',
    'I love Pasta',
    'Pasta is great'
]
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(documents)
```

```python
print('Vocabulary : ',vectorizer.get_feature_names_out())
print('\n Bow Matrix :',X.toarray())
```
