```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "love", "nlp"],
    ["nlp", "is", "amazing"],
    ["i", "love", "machine", "learning"]
]
```

```python
model = Word2Vec(
    sentences,
    vector_size=5,
    window=2,
    min_count=1
)
```

```python
print(model.wv['love'])
```
