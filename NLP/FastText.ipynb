```python
from gensim.models import FastText
```

```python
sentences = [
    ["i", "love", "nlp"],
    ["nlp", "is", "amazing"],
    ["machine", "learning", "is", "fun"]
]
```

```python
model = FastText(
    sentences,
    vector_size=5,
    window=3,
    min_count=1
)

print(model.wv['nlp'])
```
