```python
import numpy as np
```

```python
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
```

```python
arr
```

```python
q1 = np.percentile(arr,25)
```

```python
q1
```

```python
q3 = np.percentile(arr,75)
```

```python
q3
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
UF = q3 + (1.5 * IQR)
UF
```

```python
LF = q1 - (1.5 * IQR)
LF
```

```python
l = []
for i in arr :
    if i >= LF and i <= UF:
        l.append(i)
arr2 = np.array(l)
```

```python
print(arr)
print(arr2)
```
