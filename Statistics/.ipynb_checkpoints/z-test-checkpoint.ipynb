```python
import numpy as np
from scipy.stats import norm
import math
```

```python
sample = [172,174,168,169,171,173,175,170,169,172]
population_mean = 170
population_std = 3
samp_mean = np.mean(sample)
n = len(sample)
```

```python
z = (samp_mean - population_mean) / (population_std / math.sqrt(n))
```

```python
print("Z-Statistic:", z)
```

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
```

```python
data = [52,54,55,53,56,54,52,55,53,54,
        55,56,54,53,52,54,55,53,54,56,
        55,54,53,52,54,55,53,56,54,55]
```

```python
z_stat, p_value = ztest(data, value=50)
```

```python
print("Z Statistic:", z_stat)
print("P-value:", p_value)
```
