```python
from scipy.stats import chi2_contingency
```

```python
# Contingency Table
data = [
    [30, 20],
    [20, 30]
]
```

```python
chi2, p, dof, expected = chi2_contingency(data)
```

```python
print("Chi-Square Statistic:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
```
