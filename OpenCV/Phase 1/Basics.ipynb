```python
# importing opencv
import cv2
```

```python
# Check Version
print(cv2.__version__)
```

```python
# Reading an image
image = cv2.imread('Python-Symbol.png')
```

```python
# Displaying an image 
if image is None:
    print("image not found")
else:
    cv2.imshow('Python Image',image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#Saving an image 
cv2.imwrite('PythonLogo.png',image)
```

```python
# Image Dimensions
print(image.shape)
print(image.shape[0])
print(image.shape[1])
print(image.shape[2])
print(image.size)
print(image.dtype)
```

```python
# Grayscale Conversion
gray = cv2.cvtColor(
    image,
    cv2.COLOR_BGR2GRAY
)
cv2.imshow("Gray", gray)

cv2.waitKey(0)

cv2.destroyAllWindows()
```
