```python
import cv2
```

```python
img = cv2.imread("dog.png")
```

```python
# Resizing
resized = cv2.resize(img, (500, 400))

cv2.imshow("Resized", resized)
cv2.imwrite("Resized Image.png",resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Cropping using slicing
crop = img[50:250, 100:300]

cv2.imshow("Crop", crop)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotating
rotated = cv2.rotate(
    img,
    cv2.ROTATE_90_CLOCKWISE
)

cv2.imshow("Rotate", rotated)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Flipping 
flip = cv2.flip(img, -1)

cv2.imshow("Flip", flip)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Drawing Shapes
cv2.line(
    img,
    (50,50),
    (300,300),
    (255,0,0),
    3
)

cv2.rectangle(
    img,
    (100,100),
    (300,250),
    (0,255,0),
    2
)

cv2.circle(
    img,
    (250,200),
    60,
    (0,0,255),
    3
)

cv2.circle(
    img,
    (250,200),
    60,
    (255,0,0),
    -1
)

cv2.imshow("Line", img)

cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
# Adding Text
cv2.putText(
    img,
    "OpenCV",
    (100,100),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (255,255,255),
    2
)

cv2.imshow("Text", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
