```python
import cv2
import numpy as np
```

```python
image = cv2.imread('nature.png')
```

```python
# Gaussian Blue 
blur = cv2.GaussianBlur(image , (5,5) , 0)
cv2.imshow("Guassian Blur",blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Median Blur
median = cv2.medianBlur(
    image,
    5
)

cv2.imshow("Median Blur", median)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])

sharp = cv2.filter2D(
    image,
    -1,
    kernel
)

cv2.imshow("Sharpen", sharp)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
